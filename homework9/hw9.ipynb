# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
#Define functions
def f(x):
    return x**2

def g(x):
    return x**0.5

#Define variable x
x=np.linspace(0, 10, num=11, endpoint=True)

#Create figure, label axes & title
plt.figure(figsize=(10,8))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph of f(x) and g(x)')

#Line graphs of each function with different styles
plt.plot(f(x), color='blue', linestyle='--', label='f(x)')
plt.plot(g(x), color='red', linestyle=":", label='g(x)')

#Legend using labels specified above
plt.legend()

plt.show()

#Print lists of x values and corresponding f(x), g(x)
print("x=", x)
print("Values of f(x):", f(x))
print("Values of g(x):", g(x))
```

```python
#Define functions
def f(x):
    return x**2

def g(x):
    return x**0.5

#Define variable x
x=np.linspace(0, 10, num=11, endpoint=True)

#Create figure, label axes & title
plt.figure(figsize=(10,8))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph of f(x) and g(x)')

#Line graphs of each function with different styles
plt.plot(f(x), color='blue', linestyle='--', label='f(x)')
plt.plot(g(x), color='red', linestyle=":", label='g(x)')

#Plot horizontal and vertical lines
plt.axhline(y=3, color='yellow')
plt.axvline(x=3, color='yellow')

#Limit x and y axes to range 0,5
plt.xlim(0,5)
plt.ylim(0,5)

#Plot the sum of f(x) and g(x)
plt.plot(x, f(x)+g(x), color='green', label='sum')


#Legend outside figure on upper right corner
plt.legend(loc='upper right', bbox_to_anchor=(1.2,1.0))

plt.show()

#Print lists of x values and corresponding f(x), g(x)
print("x=", x)
print("Values of f(x):", f(x))
print("Values of g(x):", g(x))
print("Values of sum:", f(x)+g(x))
```
